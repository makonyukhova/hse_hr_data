# Поиск аномальных значений

Для начала определимся, какие значения в принципе могут быть названы аномальными. 

Предположим, что это могут быть: 
* Сотрудники, у которых повышение зарплаты в процентах больше, чем у 90% персонала 
* Сотрудники, обладающие уровнем месячного дохода значительно ниже или выше основной массы персонала


Классически "выбросами" считаются значения за пределами трех среднеквадратических отклонений от медианного значения в выборке. Можно быстро вывести их на экран банальным боксплотом и посмотреть, есть они вообще или нет. Обычно в таком графике все "точки" за пределами "усов" "ящика" являются выбросами.

```python
# Эффективно выделим кусок данных через numpy where

df_hr['abnormal_income'] = np.where(
                            df_hr['MonthlyIncome'] > 
                            np.mean(df_hr.MonthlyIncome) + 3*np.std(df_hr.MonthlyIncome),                 
                            1, 0)
```

```python
df_abnormal_income = df_hr[df_hr['abnormal_income'] == 1]

df_abnormal_income
```

В датасете нет сотрудников с ультравысоким месячным доходом. Вероятно, распределение зарплат в компании достаточно равномерное, и доход топ-менеджеров не является аномально высоким.

## Оценка темпов повышения зарплаты сотрудников

```python
plt.hist(df_hr['PercentSalaryHike'], color = 'orange', edgecolor = 'black', bins = 20)
```

Распределение темпов повышения уровня зарплаты смещено влево, т.е. доход наибольшего количества сотрудников вырос на 11-14%. Больший рост получило меньшее число сотрудников. 

Из метода .describe() выше нам известно, что 75-перцентиль повышения зарплаты в процентах находится на уровне 18. 

Давайте также посчитаем, где лежит 90-й перцентиль. 



```python
df_hr.PercentSalaryHike.quantile(q=0.9)
```

Это означает, что зарплаты 90% сотрудников в выборке выросли менее, чем на 21%.

Давайте объявим аномалиями все, что выше этой отметки, и выведем тех, чья зарплата выросла на 21 и более процент. 

```python
df_hr_h = df_hr[df_hr['PercentSalaryHike'] > 21]
df_hr_h
```

Получилось 123 человека, чья зарплата выросла на 21-25%.
